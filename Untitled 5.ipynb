```python
# Set the correlation threshold
threshold = 0.8

# Find feature pairs with correlation >= 80%
high_corr_pairs = []
for i in range(len(corr_matrix.columns)):
    for j in range(i + 1, len(corr_matrix.columns)):  # Avoid self-correlation
        if corr_matrix.iloc[i, j] >= threshold:
            high_corr_pairs.append((corr_matrix.columns[i], corr_matrix.columns[j], corr_matrix.iloc[i, j]))

# Print the highly correlated feature pairs
if high_corr_pairs:
    print("Feature pairs with correlation >= 80%:\n")
    for pair in high_corr_pairs:
        print(f"{pair[0]} ↔ {pair[1]}  | Correlation: {pair[2]:.2f}")
else:
    print("No highly correlated feature pairs found.")

# Decide which features to remove (keeping only one from each pair)
features_to_remove = set()
for col1, col2, _ in high_corr_pairs:
    features_to_remove.add(col2)  # Arbitrarily remove the second feature in each pair

# Print the removed features
print("\nFeatures removed due to high correlation:")
print(", ".join(features_to_remove))

# Save total number of columns before removal
total_columns_before = data1.shape[1]

# Remove duplicate feature columns
data_reduced = data1.drop(columns=features_to_remove)

# Save total number of columns after removal
total_columns_after = data_reduced.shape[1]

# Save the cleaned dataset
data_reduced.to_csv('N_R_data_Null_Mode_Median.csv', index=False)

# Print summary
print(f"\nTotal columns before removal: {total_columns_before}")
print(f"Total columns after removal: {total_columns_after}")
print("Final dataset saved as 'N_R_data_Null_Mode_median.csv' after removing redundant features.")
```

```python
# ================== IMPORTS ==================
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
import joblib
import sys
!{sys.executable} -m pip install xgboost

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import set_random_seed

# ================== SEED ==================
np.random.seed(42)
set_random_seed(42)

# ================== CONFIG ==================
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
n_samples = len(y)

# OOF predictions
oof_nn  = np.zeros(n_samples)
oof_rf  = np.zeros(n_samples)
oof_lr  = np.zeros(n_samples)
oof_xgb = np.zeros(n_samples)

# ================== NN MODEL ==================
def build_nn_model():
    cont_input = Input(shape=(4,), name='cont')   # 4 continuous
    bin_input  = Input(shape=(8,), name='bin')    # 8 binary

    x = Concatenate()([cont_input, bin_input])
    x = Dense(32, activation='relu')(x)
    x = Dense(16, activation='relu')(x)
    output = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=[cont_input, bin_input], outputs=output)
    model.compile(optimizer=Adam(0.001), loss='binary_crossentropy')

    return model

# ================== CV LOOP ==================
for fold, (train_idx, val_idx) in enumerate(skf.split(X_cont, y)):
    print(f"\n🔁 Fold {fold+1}/{n_splits}")

    # -------- Split --------
    X_cont_train, X_cont_val = X_cont[train_idx], X_cont[val_idx]
    X_bin_train,  X_bin_val  = X_bin[train_idx], X_bin[val_idx]
    y_train, y_val = y[train_idx], y[val_idx]

    # -------- Scaling (NO LEAKAGE) --------
    scaler = StandardScaler()
    X_cont_train = scaler.fit_transform(X_cont_train)
    X_cont_val   = scaler.transform(X_cont_val)
    joblib.dump(scaler, "scaler.pkl",protocol=4) # CRITICAL: Save this for the app

    # -------- Combine for SMOTE --------
    X_train_combined = np.hstack([X_cont_train, X_bin_train])
    X_val_combined   = np.hstack([X_cont_val, X_bin_val])

    # -------- SMOTE (ONLY TRAIN DATA) --------
    smote = SMOTE(random_state=42)
    X_train_res, y_train_res = smote.fit_resample(X_train_combined, y_train)

    # Split back
    X_cont_train_res = X_train_res[:, :4]
    X_bin_train_res  = X_train_res[:, 4:]

    # ================= NN =================
    nn = build_nn_model()
    nn.fit(
        [X_cont_train_res, X_bin_train_res],
        y_train_res,
        validation_data=([X_cont_val, X_bin_val], y_val),
        epochs=30,
        batch_size=32,
        verbose=0,
        callbacks=[EarlyStopping(patience=5, restore_best_weights=True)]
    )
    oof_nn[val_idx] = nn.predict([X_cont_val, X_bin_val]).flatten()

    # ================= RF =================
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X_train_res, y_train_res)
    oof_rf[val_idx] = rf.predict_proba(X_val_combined)[:, 1]

    # ================= LR =================
    lr = LogisticRegression(max_iter=1000)
    lr.fit(X_train_res, y_train_res)
    oof_lr[val_idx] = lr.predict_proba(X_val_combined)[:, 1]

    # ================= XGB =================
    xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
    xgb.fit(X_train_res, y_train_res)
    oof_xgb[val_idx] = xgb.predict_proba(X_val_combined)[:, 1]
    
    if fold == n_splits - 1:
        # Move the scaler save here so it matches the models exactly
        joblib.dump(scaler, "scaler.pkl") 
        nn.save("base_nn.keras")
        joblib.dump(rf, "base_rf.pkl")
        joblib.dump(lr, "base_lr.pkl")
        joblib.dump(xgb, "base_xgb.pkl")

# ================== METRICS ==================
def evaluate(name, y_true, y_pred):
    y_label = (y_pred > 0.5).astype(int)
    print(f"\n📊 {name} OOF Performance:")
    print(f"F1 Score : {f1_score(y_true, y_label):.3f}")
    print(f"ROC AUC  : {roc_auc_score(y_true, y_pred):.3f}")
    print(f"Accuracy : {accuracy_score(y_true, y_label):.3f}")

evaluate("Neural Net", y, oof_nn)
evaluate("Random Forest", y, oof_rf)
evaluate("Logistic Regression", y, oof_lr)
evaluate("XGBoost", y, oof_xgb)

# ================== SAVE META DATA ==================
X_meta = np.vstack([oof_nn, oof_rf, oof_lr, oof_xgb]).T

np.save("X_meta.npy", X_meta)
np.save("y_meta.npy", y)

print("\n✅ Saved: X_meta.npy, y_meta.npy")

# Save individual predictions
oof_base_preds = {
    "NeuralNet": oof_nn,
    "RandomForest": oof_rf,
    "LogReg": oof_lr,
    "XGBoost": oof_xgb
}
np.save("oof_base_preds.npy", oof_base_preds)

print("✅ Saved: oof_base_preds.npy")
```

```python
#Meta-Learner Code for Stacking (Level 2) GradientBoostingClassifier
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report
from sklearn.model_selection import cross_val_score, StratifiedKFold
import matplotlib.pyplot as plt
import seaborn as sns

# ========== Load Out-of-Fold Predictions ==========
X_meta = np.load("X_meta.npy")  # shape: (541, n_base_models)
y_meta = np.load("y_meta.npy")  # shape: (541,)

# ========== Choose Meta-Learner ==========
# Option 1: Simple and interpretable
# meta_model = LogisticRegression(max_iter=1000, solver='lbfgs')

# Option 2: More flexible learner
meta_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.05)

# ========== Cross-Validated Evaluation ==========
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

acc = cross_val_score(meta_model, X_meta, y_meta, scoring='accuracy', cv=cv)
f1  = cross_val_score(meta_model, X_meta, y_meta, scoring='f1', cv=cv)
auc = cross_val_score(meta_model, X_meta, y_meta, scoring='roc_auc', cv=cv)

print("\n📊 Meta-Learner Performance (5-Fold CV):")
print(f"Accuracy : {acc.mean():.3f} ± {acc.std():.3f}")
print(f"F1 Score : {f1.mean():.3f} ± {f1.std():.3f}")
print(f"ROC AUC  : {auc.mean():.3f} ± {auc.std():.3f}")

# ========== Train Final Meta-Model ==========
meta_model.fit(X_meta, y_meta)
y_pred = meta_model.predict(X_meta)
y_proba = meta_model.predict_proba(X_meta)[:, 1]

# ========== Final Evaluation ==========
print("\n🧾 Final Evaluation on Full Data:")
print(classification_report(y_meta, y_pred))
print(f"Accuracy : {accuracy_score(y_meta, y_pred):.3f}")
print(f"F1 Score : {f1_score(y_meta, y_pred):.3f}")
print(f"ROC AUC  : {roc_auc_score(y_meta, y_proba):.3f}")

# ========== Coefficients Plot (Only for Logistic Regression) ==========
if isinstance(meta_model, LogisticRegression):
    coef_df = pd.DataFrame({
        'Base Model': ['NeuralNet', 'RandomForest', 'LogReg', 'XGBoost'],
        'Coefficient': meta_model.coef_[0]
    })
    plt.figure(figsize=(6, 4))
    sns.barplot(x='Coefficient', y='Base Model',hue="Base Model", data=coef_df, palette='viridis',legend=False)
    plt.title("🔍 Meta-Learner Coefficients")
    plt.axvline(0, color='gray', linestyle='--')
    plt.tight_layout()
    plt.show()
# ========== Feature Importance Plot for Gradient Boosting Meta-Learner ==========
feature_names = ['NeuralNet', 'RandomForest', 'LogReg', 'XGBoost']  # Adjust if your meta features differ

importances = meta_model.feature_importances_
coef_df = pd.DataFrame({
    'Base Model': feature_names,
    'Importance': importances
}).sort_values('Importance', ascending=False)

plt.figure(figsize=(6, 4))
sns.barplot(
    x='Importance',
    y='Base Model',
    hue='Base Model',
    data=coef_df,
    palette='viridis',
    legend=False
)
plt.title("Meta-Learner Feature Importances (GB Classifier)")
plt.tight_layout()
plt.show()

# SAVE MODEL
joblib.dump(meta_model, "final_stacked_model_GB.pkl")
print("\n✅ Model saved as final_stacked_model_GB.pkl")

#SAVE FEATURE NAMES
joblib.dump(feature_names, "meta_feature_names.pkl")
print("✅ Feature names saved")

#SAVE PREDICTIONS
np.save("final_predictions.npy", y_pred)
np.save("final_probabilities.npy", y_proba)

print("✅ Predictions saved")
```

```python
from sklearn.metrics import confusion_matrix, precision_score, accuracy_score, f1_score, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from IPython.display import display, HTML

# ========== Confusion Matrix ==========
cm = confusion_matrix(y_meta, y_pred)

plt.figure(figsize=(4.2, 3.5))  # slightly smaller size
sns.heatmap(cm, annot=True, fmt='d', cmap='Pastel2', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel("Prediction")
plt.ylabel("Ground Truth")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()

# ========== Metric Summary Table ==========
metrics = {
    "Accuracy": accuracy_score(y_meta, y_pred),
    "F1 Score": f1_score(y_meta, y_pred),
    "Precision": precision_score(y_meta, y_pred),
    "ROC AUC": roc_auc_score(y_meta, y_proba)
}

metrics_df = pd.DataFrame(list(metrics.items()), columns=["Metric", "Value"])
metrics_df["Value"] = metrics_df["Value"].apply(lambda x: f"{x:.3f}")

# ========== Styled Table ==========
def highlight_rows(val):
    return ['background-color: lavenderblush' if i % 2 == 0 else 'background-color: honeydew' 
            for i in range(len(val))]

styled = metrics_df.style\
    .set_caption("Performance Evaluation Metrics")\
    .format(precision=3)\
    .apply(highlight_rows, axis=0)\
    .set_table_styles([
        {"selector": "caption", "props": [("font-size", "16px"), ("color", "#444"), ("font-weight", "bold"), ("text-align", "center")]},
        {"selector": "th", "props": [("text-align", "center"), ("padding", "8px 16px")]},
        {"selector": "td", "props": [("text-align", "center"), ("padding", "8px 20px")]},
        {"selector": "tbody td", "props": [("font-size", "14px")]}
    ], overwrite=False)

# Center the entire table in output display
display(HTML("<div style='display: flex; justify-content: center;'>"+styled.to_html()+"</div>"))
```

```python
import shap
import matplotlib.pyplot as plt

# ===================== Step 1: Define Feature Names =====================
feature_names = ['NeuralNet', 'RandomForest', 'LogReg', 'XGBoost']

# ===================== Step 2: Create Explainer =====================
explainer = shap.TreeExplainer(meta_model)

# ===================== Step 3: Compute SHAP Values =====================
shap_values = explainer.shap_values(X_meta)

# 🔥 Fix for binary classification (important)
if isinstance(shap_values, list):
    shap_values = shap_values[1]   # take positive class

# ===================== Step 4: Summary Plot (Beeswarm) =====================
plt.figure()
shap.summary_plot(shap_values, X_meta, feature_names=feature_names)

# ===================== Step 5: Feature Importance (Bar Plot) =====================
plt.figure()
shap.summary_plot(shap_values, X_meta, feature_names=feature_names, plot_type="bar")

# ===================== Step 6: Force Plot (Single Prediction) =====================
shap.initjs()
shap.force_plot(
    explainer.expected_value,
    shap_values[0],
    X_meta[0],
    feature_names=feature_names,
    matplotlib=True
)
```
